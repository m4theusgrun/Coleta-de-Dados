## 1\. Preços `outliers` de diamante

```
import seaborn as sns

data = sns.load_dataset("diamonds")
data.head()
```

```
# gráfico do exercício 1

diamonds = data

with sns.axes_style("whitegrid"):
  grafico = sns.boxplot(data=diamonds["price"])
  grafico.set_title("Preços de diamante")
  grafico.set_xlabel("Preço")
  grafico.set_ylabel("Quantidade")
  grafico.set_ylim(0, 15000)
  grafico.grid(True)
  grafico.figure.set_size_inches(10, 5)
```

**Insight do gráfico 1**: ...

```
#neste grafico podemos analisar um padrão que não estamos tão acostumados, temos como informação a quantidade e o preço do diamante, no qual se mantave dentro de uma 'box', sendo entre os numeros do 1.000(mil), chegando até a casa dos 5.000(mill)
```

## 3\. Distribuição contínua aproximada do peso (`weight`) de carros

```
import seaborn as sns

data = sns.load_dataset("mpg")
data.head()
```

```
# gráfico do exercício 3

cars = data

with sns.axes_style('whitegrid'):
  grafico = sns.histplot(data=cars, x="weight")
  grafico.set(title='Distribuição contínua aproximada do peso de carros' , xlabel='Peso', ylabel='mpg');
```

**Insight do gráfico 3**: ...

```
#neste grafico temos a analise de distribuição continua aproximada do peso de carros, onde se manteve digamos que um padrão equilibrado, com exceção para os pesos de 2000 quilos, que teve uma alta notável, e uma queda significativa a partir de 4500 a 5000 quilos.
```
